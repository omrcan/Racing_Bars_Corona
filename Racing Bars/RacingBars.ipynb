```python
import pandas as pd
```

```python
df=pd.read_csv('corona_dat.csv')
```

```python
df.sample()
```

```python
df2=df[['date','China','Italy','Brazil','Spain','US', 'Turkey']]
```

```python
df2.head()
```

```python
df2.set_index('date',inplace=True)
```

```python
sum=df2.cumsum(axis=0)
```

```python
sum.head()
```

```python
import bar_chart_race as bcr
```

```python
bcr.bar_chart_race(sum,filename='covid19.mp4', figsize=(10,8),title='Omer Can' )
```
